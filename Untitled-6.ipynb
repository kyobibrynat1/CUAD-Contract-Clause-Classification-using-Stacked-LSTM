## Problem Statement

Classifying legal contracts by clause type. Lawyers spend hours manually reading and categorizing contract clauses - this project automates that process using deep learning.

## Why Deep Learning?

Traditional methods like keyword matching don't understand context. LSTMs can read sequences and understand the meaning, making them good for legal text.

## Solution: Stacked LSTM

I'm using a 2-layer LSTM network because:
- Legal text has long-range dependencies
- Need to remember context from earlier in the document
- Stacking layers helps capture complex patterns

## Dataset

**CUAD v1** - Contract Understanding Atticus Dataset
- 500+ commercial contracts
- 13,000+ labeled clauses
- 41 different clause types

Using the top 10 most common clause types for this project.

## Head-Tail Sampling

To save memory and training time, I'm only using:
- First 1000 words (definitions, parties, scope)
- Last 1000 words (termination, signatures, final terms)

The middle is usually repetitive, so this captures the important parts.

```python
def clean_text(text):
    """Basic text cleaning"""
    if not isinstance(text, str):
        return ""
    
    text = text.lower()
    text = re.sub(r'[^a-z\s\.,;:\-]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

# Test
sample = "THIS AGREEMENT is made on January 1, 2020!!!"
print("Before:", sample)
print("After:", clean_text(sample))
```

# 5. Head-Tail Sampling

```python
def head_tail_extract(text, head_words=1000, tail_words=1000):
    """Extract first and last N words"""
    if not text:
        return ""
    
    words = text.split()
    if len(words) <= (head_words + tail_words):
        return text
    
    head = words[:head_words]
    tail = words[-tail_words:]
    return ' '.join(head + tail)

# Test
test = " ".join([f"word{i}" for i in range(2500)])
sampled = head_tail_extract(test, 100, 100)
print(f"Original: 2500 words → Sampled: {len(sampled.split())} words")
```

```python
# Apply sampling
df['sampled_text'] = df['cleaned_text'].apply(head_tail_extract)
print(f"✓ Sampled {len(df)} documents")
```

# 6. Tokenization

```python
class CustomTokenizer:
    """Simple tokenizer - built from scratch"""
    
    def __init__(self, vocab_size=10000):
        self.vocab_size = vocab_size
        self.word_to_index = {"<OOV>": 1}
        self.word_counts = Counter()
        
    def fit_on_texts(self, texts):
        for text in texts:
            self.word_counts.update(str(text).split())
        
        most_common = self.word_counts.most_common(self.vocab_size - 2)
        for idx, (word, _) in enumerate(most_common, start=2):
            self.word_to_index[word] = idx
        
        print(f"Vocabulary size: {len(self.word_to_index)}")
    
    def texts_to_sequences(self, texts):
        sequences = []
        for text in texts:
            seq = [self.word_to_index.get(word, 1) for word in str(text).split()]
            sequences.append(seq)
        return sequences
    
    def get_vocab_size(self):
        return len(self.word_to_index)

# Build vocab
tokenizer = CustomTokenizer(vocab_size=10000)
tokenizer.fit_on_texts(df['sampled_text'])
```

```python
# Convert to sequences
sequences = tokenizer.texts_to_sequences(df['sampled_text'])
print(f"Converted {len(sequences)} documents to sequences")
```

# 9. Build Model

```python
def build_model(vocab_size, max_length, num_classes, lstm_1=128, lstm_2=64, dropout=0.3):
    """Stacked LSTM model"""
    model = Sequential([
        Embedding(vocab_size + 1, 128, input_length=max_length),
        LSTM(lstm_1, return_sequences=True),
        Dropout(dropout),
        LSTM(lstm_2),
        Dropout(dropout),
        Dense(64, activation='relu'),
        Dropout(dropout),
        Dense(num_classes, activation='softmax')
    ])
    return model

VOCAB_SIZE = tokenizer.get_vocab_size()
NUM_CLASSES = len(label_encoder.classes_)

print(f"Vocab: {VOCAB_SIZE}, Classes: {NUM_CLASSES}, Max length: {MAX_LENGTH}")
```
